```python
from pyspark.sql import SparkSession

spark = SparkSession \
    .builder \
    .appName("AnaliseRegressao") \
    .getOrCreate()
```

```python
diretorioRegressao="./pagamentoSeguro.csv"
```

```python
from pyspark.sql.types import *
schema=StructType().add("X", StringType(),True).add("Y", FloatType(),True)
```

```python
pagamentoSeguro = spark.read.format('csv').schema(schema).options(header='true', delimiter=';').load(diretorioRegressao)
```

```python
pagamentoSeguro.printSchema()
```

```python
pagamentoSeguro.show()
```

```python
import pyspark.sql.functions as F
pagamentoSeguro = pagamentoSeguro.select(F.col("X").alias("numero_da_apolice"), 
                                         F.col("Y").alias("valor_cobrado"))
```

```python
pagamentoSeguro.show()
```

```python
pagamentoSeguroFinal=pagamentoSeguro.select(F.col('numero_da_apolice'),F.col('valor_cobrado'),pagamentoSeguro.numero_da_apolice.cast('float').alias('Numero_Apolice'))
```

```python
pagamentoSeguroFinal.show()
```

```python
pagamentoSeguroFinal.printSchema()
```

```python
pagamentoSeguroFinal.describe().show()
```

```python
from pyspark.ml.feature import VectorAssembler
assembler = VectorAssembler(inputCols=['Numero_Apolice'],outputCol='features')
df_seguro = assembler.transform(pagamentoSeguroFinal)
df_seguro.printSchema()
```

```python
df_seguro = df_seguro.select(['features','valor_cobrado'])
df_seguro.show()
```

CRIANDO O MODELO DE REGRESSÃO LINEAR

```python
from pyspark.ml.regression import LinearRegression
lr = LinearRegression(maxIter=20, labelCol='valor_cobrado')
lrModel = lr.fit(df_seguro)
```

```python
print(f'Intercepto: {lrModel.intercept}\nCoeficiente Angular: {lrModel.coefficients.values}')
```

```python
modelsummary = lrModel.summary

print("Variância Explicada: ", modelsummary.explainedVariance) 
print('R_2: ', modelsummary.r2) #coeficiente de determinaçao da reta - quanto que a variavel x consegue explicar da variavel y?
print('Erro médio quadrático: ',modelsummary.meanSquaredError)

modelsummary.residuals.show()
```
